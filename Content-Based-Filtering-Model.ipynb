```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./Datasets/tmdb.csv")
```

```python
df.head()
```

```python
## preparing the model

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
count_model = CountVectorizer(stop_words = 'english')
count_matrix = count_model.fit_transform(df['soup'])
```

```python
cosine_sim = cosine_similarity(count_matrix , count_matrix)
```

```python
df = df.reset_index()
```

```python
indices = pd.Series(df.index , index = df['title'])
```

```python
all_titles = [df['title'][i] for i in range(len(df['title']))]
```

```python
all_titles
```

```python
def get_recommendations(title):
    cosine_sim = cosine_similarity(count_matrix , count_matrix)
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores , key = lambda a: a[1] , reverse = True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    tit = df['title'].iloc[movie_indices]
    dat = df['release_date'].iloc[movie_indices]
    return_df = pd.DataFrame(columns = ['Title' , 'Year'])
    return_df['Title'] = tit
    return_df['Year'] = dat
    
    return return_df
```

```python
## checking the output
```

```python
pred = get_recommendations("2 Fast 2 Furious")
```
